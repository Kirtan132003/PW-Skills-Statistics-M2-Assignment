```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution (e.g., values from 1 to 6)
low, high = 1, 7  # randint in scipy is exclusive of the upper bound
x = np.arange(low, high)

# Calculate PMF (Probability Mass Function)
pmf_values = randint.pmf(x, low, high)

# Plot the PMF
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution")
plt.xticks(x)
plt.show()
```

```python
from scipy.stats import bernoulli

def bernoulli_pdf(x, p):
    if x not in [0, 1] or not (0 <= p <= 1):
        raise ValueError("x must be 0 or 1, and p must be between 0 and 1.")
    
    return bernoulli.pmf(x, p)

# Test cases
p = 0.6  # Probability of success
print(f"P(X=0): {bernoulli_pdf(0, p)}")
print(f"P(X=1): {bernoulli_pdf(1, p)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success per trial
size = 1000  # Number of samples

# Generate binomial samples
samples = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(samples, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='blue', edgecolor='black')

# Overlay the true binomial PMF
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)
plt.plot(x, pmf_values, 'ro-', label='Binomial PMF')

# Labels and title
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
a, b = 1, 10  # Range [1, 10]
size = 1000    # Number of samples

# Generate discrete uniform samples
samples = np.random.randint(a, b+1, size)

# Sort the values for CDF
sorted_samples = np.sort(samples)

# Compute CDF values
cdf_values = np.arange(1, size + 1) / size

# Plot CDF
plt.step(sorted_samples, cdf_values, where='post', label='Empirical CDF', color='blue')

# Overlay Theoretical CDF
x = np.arange(a, b+1)
theoretical_cdf = randint.cdf(x, a, b+1)
plt.plot(x, theoretical_cdf, "ro-", label="Theoretical CDF")

# Labels and title
plt.xlabel("Random Variable")
plt.ylabel("Cumulative Probability")
plt.title("CDF of a Discrete Uniform Distribution")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset (normally distributed)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std=10, Size=100

# Compute Z-scores
mean = np.mean(data)
std_dev = np.std(data)
z_scores = (data - mean) / std_dev

# Plot Z-scores
plt.figure(figsize=(10, 5))
sns.histplot(z_scores, bins=20, kde=True, color='blue', edgecolor='black', label="Z-scores")

# Mark 0 (Mean) line
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")

# Labels and title
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# Display Z-scores (first 10 for reference)
print("First 10 Z-scores:", np.round(z_scores[:10], 3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a non-normal (exponential) distribution
np.random.seed(42)
population = np.random.exponential(scale=5, size=10000)  # Mean = 5

# Function to apply CLT: Draw samples and compute means
def sample_means(sample_size, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        means.append(np.mean(sample))
    return means

# Sample sizes to test CLT
sample_sizes = [10, 30, 50, 100]
plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes, 1):
    means = sample_means(size)
    plt.subplot(2, 2, i)
    sns.histplot(means, bins=30, kde=True, color='blue', edgecolor='black')
    plt.title(f"Sample Size = {size}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Function to plot Standard Normal Distribution
def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)  # Generate values from -4 to 4
    y = stats.norm.pdf(x, loc=0, scale=1)  # Standard Normal PDF

    # Plotting
    plt.figure(figsize=(8, 5))
    sns.lineplot(x=x, y=y, color='blue', label='Standard Normal Distribution')
    plt.axvline(0, color='red', linestyle='--', label='Mean (0)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (Mean=0, Std=1)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
plot_standard_normal()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a random dataset (normal distribution)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std Dev=10

# Given data point
X = 60

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Compute Z-score
z_score = (X - mean) / std_dev

# Print result
print(f"Mean of dataset: {mean:.2f}")
print(f"Standard Deviation of dataset: {std_dev:.2f}")
print(f"Z-score of {X}: {z_score:.2f}")

# Plot the normal distribution
sns.histplot(data, bins=20, kde=True, color='skyblue', stat='density')
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
y_values = stats.norm.pdf(x_values, mean, std_dev)
plt.plot(x_values, y_values, color='red', label='Normal Distribution')

# Mark the given data point
plt.axvline(X, color='green', linestyle='dashed', label=f'X={X} (Z={z_score:.2f})')
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Z-score Calculation & Normal Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

# Step 1: Define population parameters
population_mean = 50
population_std = 10  # Assume known population standard deviation
sample_size = 30

# Step 2: Generate sample data
np.random.seed(42)
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 3: Calculate sample mean
sample_mean = np.mean(sample)

# Step 4: Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Determine critical value for a 95% confidence level
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)  # Two-tailed test

# Step 6: Hypothesis Testing Decision
if abs(z_score) > critical_value:
    decision = "Reject the null hypothesis (H0)"
else:
    decision = "Fail to reject the null hypothesis (H0)"

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"Critical Value: {critical_value:.2f}")
print(f"Decision: {decision}")
```

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate a sample dataset
np.random.seed(42)
sample_size = 30
population_mean = 100
population_std = 15  # Assume population standard deviation is unknown
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Compute sample mean and standard deviation
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Use ddof=1 for sample standard deviation

# Step 3: Calculate the confidence interval (95% confidence level)
alpha = 0.05  # 95% confidence interval
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)  # t-score for 95% confidence

margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Define the Poisson distribution parameter (λ = expected occurrences per interval)
lambda_value = 4  # Example: Average 4 events per interval

# Step 2: Generate a range of values (k) to calculate the CDF
k = np.arange(0, 15)  # Values from 0 to 14

# Step 3: Compute the CDF using Poisson distribution
cdf = stats.poisson.cdf(k, lambda_value)

# Step 4: Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(k, cdf, where='mid', color='red', linewidth=2, label=f"Poisson CDF (λ={lambda_value})")
plt.xlabel("Number of Events (k)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
plt.legend()
plt.grid()
plt.show()
```
